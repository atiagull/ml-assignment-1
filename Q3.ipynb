```python
import numpy as np
from matplotlib import pyplot as plt

y_test = 7.5
velocity = np.array([0,2,4,6,8,10])
force = np.array([0, 2.9, 14.8,39.6,74.3,119])

def v_matrix(velocity):
    columns = velocity.size
    mlist = []
    mlist.append(1)
    velocity_matrix = np.ones((6,6), dtype= 'int64')
    for i in range(6):
        for j in range(6):
            velocity_matrix[i,j] = np.power(velocity[i],j)
    
    return velocity_matrix

velocity_matrix = v_matrix(velocity)
inverse = np.linalg.inv(velocity_matrix)

w = inverse.dot(force)
print("wights : ", w)
def force_predict(w,v):
    f = w[0] + w[1]*v + w[2]*v**2 + w[3]*v**3 + w[4]*v**4 + w[5]*v**5
    return f

plt.scatter(velocity, force)
plt.plot(velocity, w[0] + w[1]*velocity + w[2]*velocity**2 + w[3]*velocity**3 + w[4]*velocity**4 + w[5]*velocity**5)
print("predicted values of 7.5 is using interpolating polynomial: ", f)
plt.scatter(y_test,f)
```
